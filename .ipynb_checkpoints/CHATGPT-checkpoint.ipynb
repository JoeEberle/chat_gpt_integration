```python
def get_AI_response(question):       # Generate a response from the OpenAI chatgpt3 API using the davinci model
    ''' Ask a question of an AI system and get a response'''
    completions = openai.Completion.create(engine=model_engine, prompt=question, max_tokens=1024, n=1, stop=None, temperature=0.5,)
    return completions.choices[0].text
```

```python
import pandas as pd 
import numpy as np 
from scipy.stats import rank 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid") 

def get_question_and_answer_df(question_and_answer_csv_file, answer_df_file): 
    df = pd.read_csv(question_and_answer_csv_file) 
    answer_df = pd.read_csv(answer_df_file) 
    df = df.join(answer_df,how='inner',left_index=True,right_index=True) 
    df.drop('Answer',axis=1,inplace=True) 
    df.drop('Question',axis=1,inplace=True) 
    return df 

question_df = get_question_and_answer_df('question_and_answer.csv', 'question_and_answer.csv') 
question_df.head() answer_df.head()
```
